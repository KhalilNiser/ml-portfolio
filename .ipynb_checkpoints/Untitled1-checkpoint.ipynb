```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt


# Create data training
x = torch.linspace(0, 10, 100).unsqueeze(1)
# Noisy line (slope = 2, intercept = 1)
y = x * 2 + 1 + torch.randn(x.size()) * 0.5


# Neural Network Model
model = nn.Sequential(
    nn.Linear(1, 16),
    nn.ReLU(),
    nn.Linear(16, 1)
)



# Loss + optimizer
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

losses = []


# Training loop
for epoch in range(300):
    pred = model(x)
    loss = loss_fn(pred, y)


    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


    losses.append(loss.item())


    # Plot results
    plt.figure()
    plt.scatter(x.detach(), y.detach())
    plt.plot(x.detach(), model(x).detach())
    plt.title("Neural Network Learned Line")
    plt.show()


    # Print last loss
    print("Final Loss:", losses[-1])
    
```
